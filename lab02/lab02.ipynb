```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
k = 19
fi = 3*math.pi/4
```

Начальные условия:
1) te0 = 0, r0 = 10/61*k
2) te0 = -pi, r0 = 10/41*k

```python
def dr(r, tetha): #функция, описывающая движение катера береговой охраны
    dr = r/math.sqrt(25.01)
    return dr
```

```python
r01 = 10/61*k #1 случай
r02 = 10/41*k #2 случай

te = np.arange(0, 2*math.pi, 0.01) #массив для тетты

r1 = odeint(dr, r01, te)
r2 = odeint(dr, r02, te)
```

```python
def xt(t): #функция, описывающая движение лодки браконьеров
    xt = math.tan(fi)*t
    return xt
```

```python
t = np.arange(0, 20, 1)
```

```python
#Перевод в полярные координаты
tete = (np.tan(xt(t)/t))**-1
rr = np.sqrt(t*t + xt(t)*xt(t))
```

```python
plt.polar(te, r1, 'g') #построение траектории движения катера в полярных координатах. 1 случай
plt.polar(tete, rr, 'b') #построение траектории движения лодки в полярных координатах
```

```python
plt.polar(te, r2, 'g') #построение траектории движения катера в полярных координатах. 2 случай
plt.polar(tete, rr, 'b') #построение траектории движения лодки в полярных координатах
```

```python
idx = np.argwhere(np.diff(np.sign(rr - r1))).flatten()
#точки пересечения для первого графика
print (tete[-1])
print (rr[idx[-1]])
```

```python
idx = np.argwhere(np.diff(np.sign(rr - r2))).flatten()
#точки пересечения для второго графика
print (tete[-1])
print (rr[idx[-1]])
```
