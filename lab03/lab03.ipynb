```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Начальные условия
x0=882000
y0=747000

a1 = 0.4
b1 = 0.67
c1 = 0.77
g1 = 0.14

a2 = 0.24
b2 = 0.67
c2 = 0.47
g2 = 0.14
# Условия описывающие время
t0 = 0
tmax = 1
dt = 0.05
t = np.arange(t0, tmax, dt)

def P1(t):
    p1 = np.sin(3*t)+1
    return p1
def Q1(t):
    q1 = np.cos(2*t)+2
    return q1

def P2(t):
    p2 = np.fabs(np.sin(2*t))
    return p2
def Q2(t):
    q2 = np.fabs(np.cos(2*t))
    return q2
# дифференциальных уравнений, описывающих изменение численности армий
def S1(f, t):
    s11 = -a1*f[0]-b1*f[1]+P1(t)
    s12 = -c1*f[0]-g1*f[1]+Q1(t)
    return s11, s12
def S2(f, t):
    s21 = -a2*f[0]-b2*f[1]+P2(t)
    s22 = -c2*f[0]*f[1]-g2*f[1]+Q2(t)
    return s21, s22
# Вектор начальной численности армий
v = np.array([x0,y0])
# Решение системы уравнений
f1 = odeint (S1, v, t)
f2 = odeint (S2, v, t)
# График
plt.plot(t, f1)
plt.ylabel('Численность армии')
plt.xlabel('Время')
plt.legend(['Армия Х', 'Армия Y'])
```
