```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

x0 = 11
N = 1130

t0 = 0
tmax = 18
dt = 0.1
t = np.arange(t0, tmax, dt)

def k1(t):
    g = 0.25
    return g
def k2(t):
    g = 0.000075
    return g
def k3(t):
    g = 0.25*np.sin(t)
    return g
def k4(t):
    g = 0.01
    return g
def p1(t):
    v = 0.000075
    return v
def p2(t):
    v = 0.25
    return v
def p3(t):
    v = 0.75*t
    return v
def p4(t):
    v = 0.001
    return v

def f1(x, t):
    xd = (k1(t) + p1(t)*x)*(N - x)
    return xd
def f2(x, t):
    xd = (k2(t) + p2(t)*x)*(N - x)
    return xd
def f3(x, t):
    xd = (k3(t) + p3(t)*x)*(N - x)
    return xd
def f4(x, t):
    xd = k4(t)*(N - x)
    return xd
def f5(x, t):
    xd = (p4(t)*x)*(N - x)
    return xd

x1 = odeint(f1, x0, t)
x2 = odeint(f2, x0, t)
x3 = odeint(f3, x0, t)
x4 = odeint(f4, x0, t)
x5 = odeint(f5, x0, t)

plt.plot(t, x1, label='Первый случай')
plt.plot(t, x2, label='Второй случай')
plt.plot(t, x3, label='Третий случай')
plt.legend()
```
